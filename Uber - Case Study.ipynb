```python
#Drop the unknown location value from START and STOP column

df_2=df[df["START"]!="Unknown Location"]
df_2=df[df["STOP"]!="Unknown Location"]
```

```python
#Find the most popular START-STOP pair according to the total number of rides covered

df_2.groupby(["START","STOP"]).size().sort_values(ascending=False).reset_index(name="Count")
```

```python
#Convert the datatypes of START_DATE and END_DATE columns to datetime

df['START_DATE'] = pd.to_datetime(df['START_DATE'])
df['END_DATE'] = pd.to_datetime(df['END_DATE'])
```

```python
df.info()
```

```python
#Extract the month from START DATE and add it to a new column named "MONTHS"

df['MONTHS'] = df['START_DATE'].dt.month
```

```python
df
```

```python
#Extract the month name from START DATE and add it to a new column named "MONTH_NAME"

df["MONTH_NAME"] = df["START_DATE"].dt.month_name()
```

```python
df
```

```python
#find the average distance covered each month

df.groupby("MONTH_NAME").mean(numeric_only=True)[["MILES"]]
```
